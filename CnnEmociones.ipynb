```python
import numpy as np
import keras
import random
from keras.preprocessing import image
from glob import glob
import cv2, os
import matplotlib.pyplot as plt
```

```python
path='train'
ROW, COL = 48, 48
```

```python
enojado2, feliz2, neutral2, sorprendido2, triste2 = [], [], [], [], [],
y_enojado2, y_feliz2, y_neutral2, y_sorprendido2, y_triste2 = [], [], [], [], []
```

```python
#Se cargan las imagenes
   
enojado_path = os.path.join(path, 'enojado2', 'im*')
for enojado_img in glob(enojado_path):
    enojado = cv2.imread(enojado_img)
    enojado = cv2.cvtColor(enojado, cv2.COLOR_BGR2GRAY)
    enojado = cv2.resize(enojado, (ROW,COL))
    enojado = image.img_to_array(enojado)
    enojado2.append(enojado)

feliz_path = os.path.join(path, 'feliz2', 'im*')
for feliz_img in glob(feliz_path):
    feliz = cv2.imread(feliz_img)
    feliz = cv2.cvtColor(feliz, cv2.COLOR_BGR2GRAY)
    feliz = cv2.resize(feliz, (ROW,COL))
    feliz = image.img_to_array(feliz)
    feliz2.append(feliz)

neutral_path = os.path.join(path, 'neutral2', 'im*')
for neutral_img in glob(neutral_path):
    neutral = cv2.imread(neutral_img)
    neutral = cv2.cvtColor(neutral, cv2.COLOR_BGR2GRAY)
    neutral = cv2.resize(neutral, (ROW,COL))
    neutral = image.img_to_array(neutral)
    neutral2.append(neutral)

sorprendido_path = os.path.join(path, 'sorprendido2', 'im*')
for sorprendido_img in glob(sorprendido_path):
    sorprendido = cv2.imread(sorprendido_img)
    sorprendido = cv2.cvtColor(sorprendido, cv2.COLOR_BGR2GRAY)
    sorprendido = cv2.resize(sorprendido, (ROW,COL))
    sorprendido = image.img_to_array(sorprendido)
    sorprendido2.append(sorprendido)

triste_path = os.path.join(path, 'triste2', 'im*')
for triste_img in glob(triste_path):
    triste = cv2.imread(triste_img)
    triste = cv2.cvtColor(triste, cv2.COLOR_BGR2GRAY)
    triste = cv2.resize(triste, (ROW,COL))
    triste = image.img_to_array(triste)
    triste2.append(triste)

print('Se han cargado todas las imagenes')
```

```python
print('Longitud array de enojo: {}\nLongitud array de felicidad: {}\nLongitud array de neutral: {}\nLongitud array de sorprendido: {}\nLongitud array de triste: {}'.format(len(enojado2),len(feliz2),len(neutral2),len(sorprendido2),len(triste2)))
```

```python
classes = ['enojado', 'feliz', 'neutral', 'sorprendido', 'triste']
```

```python
y_enojado2 = [4 for item in enumerate(enojado2)]
y_feliz2 = [3 for item in enumerate(feliz2)]
y_neutral2 = [2 for item in enumerate(neutral2)]
y_sorprendido2 = [1 for item in enumerate(sorprendido2)]
y_triste2 = [0 for item in enumerate(triste2)]
```

```python
enojado2 = np.asarray(enojado2).astype('float32')
feliz2 = np.asarray(feliz2).astype('float32')
neutral2 = np.asarray(neutral2).astype('float32')
sorprendido2 = np.asarray(sorprendido2).astype('float32')
triste2 = np.asarray(triste2).astype('float32')

y_enojado2 = np.asarray(y_enojado2).astype('int32')
y_feliz2 = np.asarray(y_feliz2).astype('int32')
y_neutral2 = np.asarray(y_neutral2).astype('int32')
y_sorprendido2 = np.asarray(y_sorprendido2).astype('int32')
y_triste2 = np.asarray(y_triste2).astype('int32')

enojado2 /= 255
feliz2 /= 255
neutral2 /= 255
sorprendido2 /= 255
triste2 /= 255
```

```python
X = np.concatenate((enojado2,feliz2,neutral2,sorprendido2,triste2), axis=0)
y = np.concatenate((y_enojado2, y_feliz2, y_neutral2, y_sorprendido2, y_triste2), axis=0)
```

```python
from keras.models import Sequential
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.layers.core import Dense, Flatten, Dropout, Activation
from keras.optimizers import Adam
from keras.utils import np_utils
from keras.callbacks import ModelCheckpoint
from keras.models import save_model, load_model
```

```python
IMG_CHANNEL = 1
BATCH_SIZE = 64
N_EPOCH = 10
VERBOSE = 2
VALIDAION_SPLIT = .2
OPTIM = Adam()
N_CLASSES = len(classes)
```

```python
y = np_utils.to_categorical(y, N_CLASSES)
print('One-Hot Encoding listo')
```

```python
model = Sequential([
    Conv2D(32, (3,3), padding='same', input_shape=(ROW, COL, IMG_CHANNEL), activation='relu'),
    Conv2D(32, (3,3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.25),
    Conv2D(64, (3,3), padding='same', activation='relu'),
    Conv2D(64, (3,3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.25),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(.5),
    Dense(N_CLASSES, activation='softmax')
])
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=OPTIM, metrics=['accuracy'])
checkpoint = ModelCheckpoint('emociones_checkpoint.h5')
```

```python
model.fit(X, y, batch_size=BATCH_SIZE, epochs=N_EPOCH, validation_split=VALIDAION_SPLIT, verbose=VERBOSE)
```

```python
scores = model.evaluate(X, y, verbose=2)
print('MODEL ACCURACY\n{}: {}%'.format(model.metrics_names[1], scores[1]*100))
```
